```python
import pandas as pd

data = pd.read_csv('wiki_languages_most_recent.csv')
```

```python
data['lang'].loc[1] = 'Tlapanec language'
data = data.drop(columns = ['Unnamed: 0'])
# data = data.set_index('lang')
data
```

```python
lang_col = data['lang']
lang_col = lang_col.str.replace('_', ' ')
lang_col = lang_col.str.replace(' language', '')
lang_col
```

```python
family_col = data['family']
family_col = family_col.apply(lambda x: [] if pd.isnull(x) else x)
family_col = family_col.apply(lambda values: [values[:x.find('[')] if '[' in x else x for x in values])

# family_col = family_col.apply(lambda values: [x[:x.find('[')] if '[' in x else x for x in values])
family_col
```
