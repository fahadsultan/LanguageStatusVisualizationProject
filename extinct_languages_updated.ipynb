```python
import pandas as pd 
import numpy as np
import requests
from tqdm import tqdm

data = pd.read_csv('Extinct languages - DATA SUMMARY.csv')
```

```python
data
```

```python
import requests
import re
languages = data['Name in English'].to_list()
wikipedia_links = []


for language in tqdm(languages):
    url = f"https://en.wikipedia.org/wiki/{language.replace(' ', '_')}"
    response = requests.get(url)
    if response.status_code == 200:
        match = re.search(r'<link rel="canonical" href="([^"]+)"', response.text)
        if match:
            wikipedia_links.append(match.group(1))
        else:
            wikipedia_links.append(None)
    else:
        wikipedia_links.append(None)

wikipedia_links
```

```python
data.loc[0:2539, 'Wikipedia URL'] = wikipedia_links
```

```python
data.to_csv('extinct_languages_links.csv', index=False)
```
